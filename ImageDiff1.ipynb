# Extract the difference between two images

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image, ImageFile
import cv2
import numpy as np
import pandas as pd
```

```python
img1 = "beach.jpg"
img2 = "beach2.jpg"
img3 = "space.jpg"

def plotit (img):
    plt.figure(figsize = (15,8))
    plt.imshow(img,cmap = 'gray')
    plt.show()
```

```python
image1 = plt.imread(img1)
plotit (image1)
```

```python
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
plotit (image1)
```

```python
image2 = plt.imread(img2)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
plotit (image2)
```

```python
imageDiff = cv2.subtract (image1, image2)
plotit (imageDiff)
```

```python
imageDiffInverted = cv2.bitwise_not(imageDiff)
plotit (imageDiffInverted)
```

```python
background = plt.imread(img3)
background = cv2.cvtColor(background, cv2.COLOR_BGR2GRAY)
backgroundInverted = cv2.bitwise_not(background)
plotit (backgroundInverted)
```

```python
imageAdd = cv2.addWeighted(backgroundInverted,0.4,imageDiff,1.0,0)
imageAddInverted = cv2.bitwise_not(imageAdd)
plotit (imageAddInverted)
```
